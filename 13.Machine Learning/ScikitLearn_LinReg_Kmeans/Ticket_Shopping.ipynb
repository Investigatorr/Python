```python
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('Ticket,Shoppint.txt', sep = ',')
df.head()
```

```python
df.corr()
```

```python
df2 = df.copy()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
mglearn.discrete_scatter(df2.iloc[:,0], df2.iloc[:,1], kmeans.labels_)
plt.legend(['Cluster 0 ', 'Cluster 1', 'Cluster 2'], loc = 'best')
plt.xlabel('Ko')
plt.ylabel('Eng')
plt.show()
```

```python
res_temp = kmeans.labels_
res_temp
```

```python
df2 = pd.DataFrame(res_temp, columns = ['Classification'])
```

```python
pd.concat([df, df2], axis = 1)
```
