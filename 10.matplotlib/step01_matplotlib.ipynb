```python
plt.plot(data, data, marker = 'o', ls = '-')
plt.plot(data, data+1, marker = 'o', ls = '--')
plt.plot(data, data+2, marker = 'o', ls = '-.')
plt.plot(data, data+3, marker = 'o', ls = ':')
```

### x / y축 끝점 설정

```python
plt.plot(data, data, marker = 'o', ls = '-')
plt.plot(data, data+1, marker = 'o', ls = '--')
plt.plot(data, data+2, marker = 'o', ls = '-.')
plt.plot(data, data+3, marker = 'o', ls = ':')

# x축 0~5 / y축 1~5까지 제한
plt.xlim(0,5)  
plt.ylim(1,5)
plt.show()
```

```python
plt.plot(data, data, marker = 'o', ls = '-')
plt.plot(data, data+1, marker = 'o', ls = '--')
plt.plot(data, data+2, marker = 'o', ls = '-.')
plt.plot(data, data+3, marker = 'o', ls = ':')

# title 설정
plt.title('data view')

# x, y label => 라벨이 있어야 뒤에 plt.legen()로 범례를 나눠 부여할 수 있다.
plt.xlabel('x data')
plt.ylabel('y data')

plt.xlim(0,5)  
plt.ylim(1,5)
plt.show()
```

```python
markers = [',', '+', '.', 'o', '*',',', '+', '.', 'o', '*',',', '+', '.', 'o', '*',',', '+', '.', 'o', '*',',', '+', '.', 'o', '*',',', '+', '.', 'o', '*',',', '+', '.', 'o', '*',',', '+', '.', 'o', '*']
data = np.random.RandomState(1)
rdata = data.rand(50)

for i in range(10):
    plt.plot(rdata[i-1], rdata[i], label = i, marker = markers[i])
plt.legend()
plt.show()
```

```python
rng = np.random.RandomState() # 이 설정으로 인해 test 데이터를 동일하게 처리 가능
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

#alpha=0.3  겹치는 효과를 위한 투명도
#cmap='viridis' : 생략시 흑색
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis') # 알파가 투명도?
plt.colorbar();  # 색상 척도 표시
```
